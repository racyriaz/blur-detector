# Import statements

```python
# for image processing
import cv2
# most essential library to handel datas and arrays
import numpy as np
# used to handel our data with ease
import pandas as pd
# display plots/images
import matplotlib.pyplot as plt
from matplotlib.image import imread
# to read files from directory 
import os
# to save train,test as pickle file data for future use
import pickle
# for visualization of datas
import seaborn as sns
# for importing Machine Learning models
from sklearn import svm
from sklearn.linear_model import LogisticRegression
# to split our X,y into training set and test set
from sklearn.model_selection import train_test_split
# to evaluate our model
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

Loading the filepaths after unzipping the folder
> make sure you change only the filepath variable below

```python
filepath = r'E:/CERTH_ImageBlurDataset/CERTH_ImageBlurDataset/'
test_path = filepath+'EvaluationSet/'
d_file = test_path+'/DigitalBlurSet.xlsx'
n_file = test_path+'/NaturalBlurSet.xlsx'

d_blur_test = test_path+'DigitalBlurSet/'
nat_blur_test = test_path+'NaturalBlurSet/'

digital_blur = pd.read_excel(d_file,names=['name','labels'])
digital_blur['name'] = digital_blur.name.apply(lambda x : x.strip()) # to remove unwanted space in file names
natural_blur = pd.read_excel(n_file,names=['name','labels'])
natural_blur['name'] = natural_blur.name.apply(lambda x : x.strip())
```

# Feature Extraction function

```python
input_size = (128, 128)
def find_features(image_path):
    image = cv2.resize(cv2.imread(image_path, cv2.IMREAD_GRAYSCALE), input_size)
    sobel_variance = np.round(cv2.Sobel(image,cv2.CV_64F,1,1,ksize=5).var(),decimals=1)
    sobel_max = np.round(cv2.Sobel(image,cv2.CV_64F,1,1,ksize=5).max(),decimals=1)
    laplace_variance = np.round(cv2.Laplacian(image, cv2.CV_64F,ksize=5).var(),decimals=1)    
    laplace_variance = np.round(cv2.Laplacian(image, cv2.CV_64F,ksize=5).max(),decimals=1)    
    return sobel_variance, sobel_max,laplace_variance, laplace_variance
```

```python
# initialization
sv_list = []
sm_list = []
lv_list = []
lm_list = []
y_true = []
```

# Loading the Input datas
Reading the image files from the directory and passing it with feature extraction function

```python
folderpath = d_blur_test
for filename in os.listdir(folderpath):
    imagepath = folderpath + filename
    sv,sm,lv,lm = find_features(imagepath)
    sv_list.append(sv)
    sm_list.append(sm)
    lv_list.append(lv)
    lm_list.append(lm)
    blur = digital_blur[digital_blur['name'] == filename].iloc[0]['labels']
    if blur == 1:
        y_true.append(1)
    else:
        y_true.append(0)
        
print("Finished loading and applied laplace varience on the filepath ... "+folderpath)
```

```python
folderpath = nat_blur_test
for filename in os.listdir(folderpath):
    imagepath = folderpath + filename
    sv,sm,lv,lm = find_features(imagepath)
    sv_list.append(sv)
    sm_list.append(sm)
    lv_list.append(lv)
    lm_list.append(lm)
    blur = natural_blur[natural_blur['name'] == filename.split('.')[0]].iloc[0]['labels']
    if blur == 1:
        y_true.append(1)
    else:
        y_true.append(0)
        
print("Finished loading and applied laplace varience on the filepath ... "+folderpath)
```

Converting the Extracted feature into DataFrame

```python
sv = pd.Series(sv_list)
sm = pd.Series(sm_list)
lv = pd.Series(lv_list)
lm = pd.Series(lm_list)
y_test = pd.Series(y_true)
features =pd.DataFrame()
features = pd.concat([sv,sm,lv,lm,y_test],axis=1)
features.columns = ['sv','sm','lv','lm','y_test']
```

# Exploratory Data Analysis with extracted features

```python
plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
plt.xlabel('sobel_variance')
plt.ylabel('sobel_max')

plt.title("sobel_variance VS sobel_maximum")
sns.scatterplot(x=features.sv,y=features.sm,hue=y_test,edgecolor='None')
plt.subplot(1,2,2)

plt.xlabel('laplace_variance')
plt.ylabel('laplace_max')
plt.title("laplace_variance VS laplace_maximum")
sns.scatterplot(x=features.lv,y=features.lm,hue=y_test,edgecolor='None')
plt.suptitle('0 -> Clear Image, 1 -> Blured image')
```

```python
# correlation of features
sns.heatmap(features.corr(),annot=True)
```

# Spliting the data into X and y terms

```python
X = features.drop(['y_test'],axis=1)
y = features['y_test']
```

Seperating the data into two categories 
> Training set (used for training the model) 
>> Test set (used for Evaluation)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# SVM model

```python
model = svm.SVC(C=100,kernel='rbf',verbose=True,)
model.fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
```

# SVM model evaluation

```python
print("Confusion Matrix: ")
print(confusion_matrix(y_test,pred))
print('\n \n')
print("Classification Report: ")
print(classification_report(y_test,pred))
print('\n \n')
print("Accuracy: ")
print(accuracy_score(y_test,pred))
```

#  Logistic Regression model

```python
logistic_model = LogisticRegression(verbose=1)
logistic_model.fit(X_train,y_train)
```

```python
l_pred = logistic_model.predict(X_test)
```

# Logistic-regression model evaluation

```python
print("Confusion Matrix: ")
print(confusion_matrix(y_test,l_pred))
print('\n \n')
print("Classification Report: ")
print(classification_report(y_test,l_pred))
print('\n \n')
print("Accuracy: ")
print(accuracy_score(y_test,l_pred))
```

saving the model, X_train,Y_train for future use

```python
#to save svm model
with open('SVM_model.pkl', 'wb') as picklefile:
    pickle.dump(model, picklefile)
with open('X_train_svm.pkl', 'wb') as picklefile:
    pickle.dump(X_train, picklefile)
with open('y_train_svm.pkl', 'wb') as picklefile:
    pickle.dump(y_train, picklefile)
```
