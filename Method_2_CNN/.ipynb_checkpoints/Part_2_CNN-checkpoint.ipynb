# Continuation from Part_1_CNN.ipynb

Import Statements

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import pandas as pd
import seaborn as sns
from matplotlib.image import imread
import os

#to save X and y
import pickle

# used for preprocessing the data
from tensorflow.keras.preprocessing import image


#used for model building
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D,MaxPool2D,Dropout,Flatten
from tensorflow.keras.callbacks import EarlyStopping
#from tensorflow.keras.optimizers import SGD
from tensorflow.keras.utils import to_categorical

#used for performance evaluation
from sklearn.metrics import classification_report,confusion_matrix

# to save our model
from tensorflow.keras.models import load_model
```

Setting same random seed as used in Part 1

```python
np.random.seed(seed=1)
tf.random.set_seed(1)
```

# Importing the saved Pickle files ( X_train, X_test, y_train, y_test)

```python
X_train = []
y_train = []
X_test = []
y_test = []
```

```python
with open('X_train.pkl', 'rb') as picklefile:
    X_train = pickle.load( picklefile)


with open('y_train.pkl', 'rb') as picklefile:
    y_train = pickle.load( picklefile)


with open('X_test.pkl', 'rb') as picklefile:
    X_test = pickle.load(picklefile)


with open('y_test.pkl', 'rb') as picklefile:
    y_test = pickle.load(picklefile)
```

Visualizing y_train and y_test

```python
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
plt.title("Y_train_dataset")
sns.countplot(x=y_train)
plt.subplot(1,2,2)
plt.title("Y_test_dataset")
sns.countplot(x=y_test)
plt.suptitle(" 0 -> Clear images , 1 -> Blurred images")
```

# Building the CNN model

```python
input_size = X_train[0].shape
```

```python
model = Sequential()
#input layer
model.add(Conv2D(filters=32,kernel_size=(5,5),activation ='relu', input_shape=(input_size[0], input_size[1], 3)))
model.add(MaxPool2D(pool_size=(2, 2)))
#hidden layer 1
model.add(Conv2D(64, (5,5),activation ='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
#hidden layer 2
model.add(Flatten())
#hidden layer 3
model.add(Dense(1024,activation='relu'))
model.add(Dropout(0.5))
#hidden layer 4
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.5))
#hidden layer 5
model.add(Dense(256,activation='relu'))
model.add(Dropout(0.5))
#hidden layer 6
model.add(Dense(30,activation='relu'))
#model.add(Dropout(0.5))
#last output layer
model.add(Dense(2,activation='softmax')) #incase  of multiclass use softmax activation function 

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy']) #loss = categorical_crossentropy for  multiclass problems
```

```python
model.summary()
```

One hot encoding the y variables

```python
traindata = np.stack(X_train)
testdata = np.stack(X_test)
trainlabel = to_categorical(y_train)
testlabel = to_categorical(y_test)
```

Inorder to prevent overfitting our model, I have made use of callback statements

```python
early_stop = EarlyStopping(monitor='val_loss',patience=2)
model.fit(x=traindata,y=trainlabel, batch_size=128,
          epochs=20,validation_data=(testdata, testlabel),callbacks=[early_stop], verbose=1)
```

```python
loss = pd.DataFrame(model.history.history)
```

# Visualizing model loss and accuracy with validation data

```python
plt.figure(figsize=(10,8))
plt.plot(loss.index,loss.loss,color='r',marker='o',linestyle='dashed')
plt.plot(loss.index,loss.val_loss,color='Blue',marker='v')
plt.title("Loss VS Validation loss",fontsize=30)
plt.gca().legend(('Loss','Validation_loss'))
plt.xlabel('epochs')
plt.ylabel('Loss')
plt.grid()
```

```python
plt.figure(figsize=(10,8))
plt.plot(loss.index,loss.accuracy,color='Black',marker='o',linestyle='dashed')
plt.plot(loss.index,loss.val_accuracy,color='Pink',marker='v')
plt.title("Accuracy VS Validation accuracy",fontsize=30)
plt.gca().legend(('Accuracy','Validation_accuracy'))
plt.xlabel('epochs')
plt.ylabel('Accuracy')
plt.grid()
```

# model accuracy

```python
(loss, accuracy) = model.evaluate(testdata, testlabel, batch_size = 128, verbose = 1)
print("accuracy: {:.2f}%".format(accuracy * 100))
```

```python
predictions = model.predict(testdata)
# accepting prediction with a chance of having 65% probability
pred = predictions > 0.65
#pred
```

```python
print(classification_report(testlabel,pred))
```
